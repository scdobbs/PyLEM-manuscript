```python
from pylem import build_model_dzdt, randomized_grid
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pylab as plt
from pyas import area, length
from scipy.stats import linregress
```

```python
dx = 100
nx = 400
ny = 200
```

```python
z0_05 = randomized_grid((ny, nx), slope = 5E-4*dx)
a, s = area(z0_05, dx)
l = length(z0_05, dx)
```

```python
fig = plt.figure()
fig.set_size_inches(18, 12)

i = np.where(np.logical_and(l.flatten() != 0, a.flatten() != 0))
plt.loglog(l.flatten()[i], a.flatten()[i],'k.')

slope, intercept, r_value, p_value, std_err = linregress(np.log10(l.flatten()[i]),np.log10(a.flatten()[i]))
print(slope, np.power(10, intercept))
```
