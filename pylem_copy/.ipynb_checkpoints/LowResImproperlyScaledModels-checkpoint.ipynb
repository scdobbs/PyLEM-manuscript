```python
from pylem import build_model_dzdt, randomized_grid
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pylab as plt

dx = 100
nx = 200
ny = 200

l = 50.0
L = 10000.0
Rf = 1000.0
time_to_steady_state = 2.0E6
Pe = 3.0
```

```python
z0_005 = randomized_grid((ny, nx), slope = 5E-5*dx)
z0_05 = randomized_grid((ny, nx), slope = 5E-4*dx)
z0_5 = randomized_grid((ny, nx), slope = 5E-3*dx)
```

# Concavity = 0.5

```python
m = 0.5
ka = 9.17
h = 1.47
```

```python
from utils import Checkpointer
checkpointer = Checkpointer('test')

f_dzdt = build_model_dzdt((ny, nx), dx, l, L, Rf, time_to_steady_state, Pe, ka, h, m, hook = checkpointer)
```

```python
results = solve_ivp(f_dzdt, y0=np.reshape(z0_005, (ny*nx,)), t_span =(0.0, time_to_steady_state*1.5), method='DOP853')
'''

dt0 = 500
t = 0
y = np.reshape(z0_005, (ny*nx,))
while t < time_to_steady_state:
    dt = dt0 / (1.0 - (t / time_to_steady_state))
    y += dt*f_dzdt(t,y)
    t += dt
'''

final = np.reshape(results.y[:,-1], (ny, nx))
fig = plt.figure()
fig.set_size_inches(18, 12)
plt.title('Concavity 0.5, 0.05% initial relief.')
plt.imshow(final)
plt.colorbar()
```

```python
from pyas import area_dinf

area, slope = area_dinf(final, dx)
fig = plt.figure()
fig.set_size_inches(18, 12)
plt.title('Concavity 0.5, 0.05% initial relief.')
plt.imshow(np.log10(area))
plt.colorbar()
```
