```python
"""
Plots bined length vs average laplacian value 
Figure 4, Dobbs and Hilley 
"""
import numpy as np
import matplotlib.pyplot as plt
import pickle as p
from pylem.pyas import length as length_calc
from pylem.pyas import area_dinf as area
from scipy.ndimage import laplace

# Define the concentrations to plot
concavities = ['01', '03', '05']
line_styles = ['-', '--', '-.']
colors = ['black', 'dimgray', 'darkgray']

plt.figure(figsize=(8, 4))

for idx, conc in enumerate(concavities):
    # Load the data
    (t, y, checkpointer) = p.load(open(f'path/to/lems/ + conc_{conc}_Rf05percent_checkpoint.p', 'rb'))
    y = np.reshape(y, (2000, 4000))

    # Calculate lengths and curvature
    lengths = length_calc(y, 10.0)
    curv = laplace(y) / 20

    # Bin the lengths and calculate mean laplacian
    binned_length_linear = np.arange(0.0, 300.0, 10.0)
    mean_curv_linear = np.zeros_like(binned_length_linear)

    for i, bll in enumerate(binned_length_linear):
        j = np.where(np.logical_and(lengths.flatten() < bll + 5.0, lengths.flatten() >= bll - 5.0))
        mean_curv_linear[i] = np.mean(curv.flatten()[j])

    # Plot the data for this concavity
    plt.plot(binned_length_linear, mean_curv_linear, label=f'Conc {conc}', color=colors[idx], linestyle=line_styles[idx])

plt.ylim(-.1, .1)
plt.xlim(0, 250)
plt.plot([0, 300], [0, 0], 'g--')  # Zero line
plt.plot([30, 30], [-1, 1], 'r--')  # Reference line
plt.xlabel('Length')
plt.ylabel('Laplacian')
plt.legend()

plt.show()
```
