```python
from TopoAnalysis import dem as d
import pandas as pd
import pickle as p
from tqdm import tqdm
import pickle
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

concavities = ["01", "03", "05"]
file_path = '/Users/scdobbs/Desktop/Folders/Active_manuscripts/Geomorph_project/Important_for_manuscript/paired_flow_lengths/junc_filtered_total_flow_lengths_combined_with_rr.csv'  # Update this with the correct path
data = pd.read_csv(file_path)
vertical_lines = {
    '01': [1, 2.2],
    '03': [1, 3],
    '05': [1, 3.17]}
                   
# Assuming `df` is your DataFrame with all necessary data
def scramble_pairs(df, concavities):
    for concavity in concavities:
        col_1 = f"total_length_1_{concavity}"
        col_2 = f"total_length_2_{concavity}"
        tl_pairs = zip(df[col_1], df[col_2])
        tl_pairs_scrambled = [np.random.choice(p, size=(2,), replace=False).tolist() for p in tl_pairs]
        df[col_1] = np.array(list(zip(*tl_pairs_scrambled))[0])
        df[col_2] = np.array(list(zip(*tl_pairs_scrambled))[1])
    return df

# Scramble pairs for all specified concavity cases
data = scramble_pairs(data, concavities)
```

```python
# Function to save checkpoint
def save_checkpoint(data, filename):
    with open(filename, 'wb') as f:
        p.dump(data, f)

# Function to load checkpoint if exists
def load_checkpoint(filename):
    if os.path.exists(filename):
        with open(filename, 'rb') as f:
            return pickle.load(f)
    else:
        return {}

# Assuming the checkpoint file path is correctly specified
# checkpoint_filename = '/Users/scdobbs/Desktop/Folders/Active_manuscripts/Geomorph_project/Important_for_manuscript/topography/Rush_run/checkpoint_junc_filter_rush_run.pkl'  # Update this to your actual file path
checkpoint_filename = '/Users/scdobbs/Desktop/Folders/Active_manuscripts/Geomorph_project/Important_for_manuscript/lems/checkpoint_conc_01.pkl'

# Load the data
filtered_results = load_checkpoint(checkpoint_filename)

# Check if data was loaded successfully
if filtered_results:
    print("Data loaded successfully.")
else:
    print("Data could not be loaded.")
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from matplotlib.colors import TwoSlopeNorm
from matplotlib.cm import ScalarMappable

# Assuming 'data' and 'concavities' are already loaded and processed

def custom_joint_plot_with_bins_and_pvalues(df, concavities, min_count=0):
    sns.set_theme(style="ticks")
    for concavity in concavities:
        min_length = np.minimum(df[f"total_length_1_{concavity}"], df[f"total_length_2_{concavity}"]) / 30
        residual = (df[f"total_length_1_{concavity}"] - df[f"total_length_2_{concavity}"]) / np.mean([df[f"total_length_1_{concavity}"], df[f"total_length_2_{concavity}"]], axis=0)
        
        plot_data = pd.DataFrame({
            'Min Length': min_length,
            'Residual': residual
        })
        
        g = sns.JointGrid(data=plot_data, x='Min Length', y='Residual', marginal_ticks=True)
        g.ax_joint.set(xscale="log")
        
        # Create an inset legend for the histogram colorbar
        cax = g.figure.add_axes([0.7, 0.55, 0.02, 0.2])         
        g.plot_joint(
            sns.histplot, bins=40, pmax=.8, cbar=True, cbar_ax=cax,
            cmap="light:#03012d", alpha=0.6
        )
        
        # Calculate bin information
        bins_global = np.logspace(np.log10(min_length.min()), np.log10(min_length.max()), 51)
        bin_means, bin_edges, _ = stats.binned_statistic(min_length, residual, statistic='mean', bins=bins_global)
        bin_stds, _, _ = stats.binned_statistic(min_length, residual, statistic='std', bins=bins_global)
        bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
        
        p_values = []
        for j in range(len(bin_centers)-1):
            bin_mask_1 = (min_length >= bin_edges[j]) & (min_length < bin_edges[j+1])
            bin_mask_2 = (min_length >= bin_edges[j+1]) & (min_length < bin_edges[j+2]) if j+2 < len(bin_edges) else (min_length >= bin_edges[j+1])
            count_bin_1 = np.sum(bin_mask_1)
            count_bin_2 = np.sum(bin_mask_2)

            if count_bin_1 > min_count and count_bin_2 > min_count:
                stat, p_value = stats.levene(residual[bin_mask_1], residual[bin_mask_2])
                p_values.append(p_value)
            else:
                p_values.append(np.nan)

        # Define a colormap for the p-values
        norm = TwoSlopeNorm(vmin=0, vcenter=0.05, vmax=1)
        cmap = plt.get_cmap('coolwarm')

        for j, p in enumerate(p_values):
            if j < len(bin_centers) - 1 and not np.isnan(p):
                color = cmap(norm(p))
                g.ax_joint.errorbar(bin_centers[j], bin_means[j], yerr=bin_stds[j], fmt='o', color=color, markersize=4, linewidth=1, mec='black', mew=0.65)
        
        g.plot_marginals(sns.histplot, element="step", color="#03012d")
        
        plt.subplots_adjust(top=0.95)
        g.fig.suptitle(f"Joint Plot of Min Length vs. Residual for Concavity {concavity}")
        plt.show()
concavities = ['01']

custom_joint_plot_with_bins_and_pvalues(data, concavities)
```
