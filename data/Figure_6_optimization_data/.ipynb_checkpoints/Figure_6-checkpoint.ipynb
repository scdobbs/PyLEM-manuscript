```python
"""
This function compiles all optimization results and plots them as 1/dx vs. stdev(ks-obs - ks-model)/ks-model
Note that it requires both pylem and TopoAnalysis to be installed onto your computer in order to load the files and calculate the area and steepnesses of the LEMs

"""

def automated_process_multiple_concs(conc_values, directory):
    # Constants and erosion coefficients
    time_to_steady_state = 2E6
    l = 30.0
    L = 10000.0
    Rf = 1E3
    m_values = {'conc0.1': 0.1, 'conc0.3': 0.3, 'conc0.5': 0.5}
    ka_values = {'conc0.1': 0.379, 'conc0.3': 5.146, 'conc0.5': 9.17}
    h_values = {'conc0.1': 1.895, 'conc0.3': 1.558, 'conc0.5': 1.47}
    scaling_factors = {'conc0.1': 1.7081925, 'conc0.3': 3.0823613, 'conc0.5': 4.1360085}

    results = {}
    plt.figure(figsize=(10, 6))
    
    for conc_value in conc_values:
        # Extract dx values for the specified conc_value from filenames in the directory
        filenames = os.listdir(directory)
        dx_values = sorted(set([float(re.search(f'{conc_value.replace(".", "_")}_dx_(\d+.\d+).p', file).group(1)) 
                                for file in filenames if re.search(f'{conc_value.replace(".", "_")}_dx_(\d+.\d+).p', file)]))
        m = m_values[conc_value]
        ka = ka_values[conc_value]
        h = h_values[conc_value]
        scaling_factor = scaling_factors[conc_value]

        K = np.power(time_to_steady_state, -1.0) * np.power(ka, -m) * np.power(1 - h * m, -1) * (
                np.power(L, 1 - h * m) - np.power(l, 1 - h * m))  # Calculate K
        K = K * scaling_factor

        U = Rf * K * np.power(ka, m) * (1 - h * m) * np.power((np.power(L, 1 - h * m) - np.power(l, 1 - h * m)), -1)
        ks = (U / K)
        print(str(conc_value) + " = "+ str(ks))
        for dx in dx_values:
            # Load file by converting dx to a string
            elev = p.load(open(directory + conc_value.replace('.', '_') + '_dx_' + str(dx) + '.p', 'rb'))
            # elev = d.FilledElevation(elevation=elev)
            grid = elev._griddata
            grid = np.pad(grid, ((1, 1), (0, 0)), 'constant', constant_values=0)

            m_size = elev._georef_info.ny+2
            n_size = elev._georef_info.nx

            # Perform the calculations
            a, s = area_calc(grid, dx=dx)
            ks_obs = np.reshape(np.power(a, m) * s, (m_size * n_size,))
            sse = np.sum(np.power(ks - ks_obs, 2))
            resid = ks_obs - ks

            # Store the results
            if conc_value not in results:
                results[conc_value] = {}
            results[conc_value][dx] = {
                'ks_obs': ks_obs,
                'resid': resid,
                'std': np.std(resid),
                'rmse': np.sqrt(sse / (m_size * n_size))

            }
        
        # Plotting for the current conc_value
        dx_values = list(results[conc_value].keys())
        normalized_std_values = [results[conc_value][dx]['std'] / ks for dx in dx_values]
        normalized_rmse_values = [results[conc_value][dx]['rmse'] / ks for dx in dx_values]
        rmse_values = [results[conc_value][dx]['rmse'] for dx in dx_values]
        plt.plot([1/dx for dx in dx_values], normalized_std_values, '.',label=conc_value)

    # Finalize the plot
    plt.xlabel('Inverse of dx')
    plt.ylabel('std(ks-obs - ks-mod)/ks-mod')
    plt.title('Inverse of dx vs Normalized Standard Deviation')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    flattened_data = []
    for conc, dx_data in results.items():
        for dx, metrics in dx_data.items():
            flattened_data.append({
                'Concavity': conc,
                'Dx': dx,
                'Residuals': metrics['resid'],
                'Standard Deviation': metrics['std'],
                'Normalized Std Deviation': metrics['std'] / ks
            })
    

    return results

# Usage:
conc_values = ['conc0.5','conc0.3', 'conc0.1']
results = automated_process_multiple_concs(conc_values, directory)
```

```python
"""
These functions perform an envelope analysis on the data in order to filter out local minima obscuring the fit of the data. 

"""

import os
import re
import numpy as np
import matplotlib.pyplot as plt
import pickle as p
from scipy.optimize import curve_fit
import seaborn as sns

from scipy.optimize import curve_fit
from sklearn.metrics import r2_score

def fit_and_evaluate_results(results, ks_mod_dict):
    def inv_exp_decay_func(x, a, b, c):
        return a * np.exp(-b * x) + c

    
    fit_parameters = {}
    r2_scores = {}

    plt.figure(figsize=(10, 6))
    colors = ['black', 'orange', 'green']
    
    for index, (conc, dx_data) in enumerate(results.items()):
        dx_values = [dx for dx in dx_data]
        if not dx_values:
            print(f"Skipping {conc} due to lack of valid data")
            continue

        inverse_dx_values = 1 / np.array(dx_values)
        normalized_std_values = np.array([dx_data[dx]['std'] for dx in dx_values])

        bounds = (0, [np.inf, np.inf, np.inf])
        try:
            popt, pcov = curve_fit(inv_exp_decay_func, inverse_dx_values, normalized_std_values, bounds=bounds, maxfev=10000)
            fit_parameters[conc] = {'a': popt[0], 'b': popt[1], 'c': popt[2]}

            predictions = inv_exp_decay_func(inverse_dx_values, *popt)
            r2_scores[conc] = r2_score(normalized_std_values, predictions)

            inverse_dx_fit = np.linspace(min(inverse_dx_values), max(inverse_dx_values), 100)
            ks_fit = inv_exp_decay_func(inverse_dx_fit, *popt)
            plt.plot(inverse_dx_fit, ks_fit, label=f'{conc} fit', color=colors[index % len(colors)])
            plt.plot(inverse_dx_values, normalized_std_values, 'o', label=f'{conc} data', color=colors[index % len(colors)], markersize=4)
        except Exception as e:
            fit_parameters[conc] = {'error': str(e)}
            r2_scores[conc] = None

    plt.xlabel('1/dx')
    plt.ylabel('Stdev(abs(residual))/ks')
    plt.title('1/dx vs Normalized Standard Deviation with Exponential Decay Fit')
    plt.semilogy()
    plt.ylim(1.5E-5, 1.0)
    plt.xlim(0, 0.005)
    plt.axhline(y=1E-4, color='r', linestyle='-')
    if plt.gca().get_legend_handles_labels()[0]:
        plt.legend()
    # plt.grid(True, which="both", ls="--", linewidth=0.5)
    plt.tight_layout()
    # Show the plot

    
    for conc_value in conc_values:
        if 'c' in fit_parameters[conc_value]:
            c_param = fit_parameters[conc_value]['c']  # Retrieve the 'c' parameter
            plt.axhline(y=c_param, color='grey', linestyle='--', label=f'Asymptote for {conc_value}')
        else:
            print(f"No fitting parameters found for {conc_value}")
            

    for conc in ks_mod_dict.keys():
        if conc in r2_scores:
            if r2_scores[conc] is not None:
                print(f'Goodness of fit for {conc}: R^2 = {r2_scores[conc]:.4f}')
            else:
                print(f'Could not compute R^2 for {conc} due to {fit_parameters[conc].get("error", "unknown error")}')
        else:
            print(f'No data for {conc}, skipped in analysis')
            
    plt.savefig('/Users/scdobbs/Desktop/Folders/Active_manuscripts/Geomorph_project/Important_for_manuscript/figures/' 
            + f"opt_results_regress_{conc}_asymptote.svg", format="svg")

    plt.show()

    return fit_parameters, r2_scores


def lower_envelope(values):
    envelope = []
    values.sort(key=lambda x: x[0], reverse=True)  # Sort by decreasing dx
    current_min = float('inf')
    for dx, y in values:
        if y < current_min:
            current_min = y
            envelope.append((dx, y))
    return envelope


def decay_function(x, a, b, c):
    return a * np.exp(-b * x) + c


def automated_process_and_plot(conc_values, directory):
    sns.set(style="whitegrid")  # Set seaborn style

    markers = ['o', 's', '^']  # Different markers for different conc cases
    colors = sns.color_palette("colorblind", len(conc_values))
    # Constants and erosion coefficients (same as in your snippets)
    time_to_steady_state = 2E6
    l = 30.0
    L = 10000.0
    Rf = 1E3
    m_values = {'conc0.1': 0.1, 'conc0.3': 0.3, 'conc0.5': 0.5}
    ka_values = {'conc0.1': 0.379, 'conc0.3': 5.146, 'conc0.5': 9.17}
    h_values = {'conc0.1': 1.895, 'conc0.3': 1.558, 'conc0.5': 1.47}
    scaling_factors = {'conc0.1': 1.7081925, 'conc0.3': 3.0823613, 'conc0.5': 4.1360085}
    
    results = {}
    plt.figure(figsize=(10, 6))

    for index, conc_value in enumerate(conc_values):
        print(f"Processing {conc_value}")

        # Extract dx values for the specified conc_value from filenames in the directory
        filenames = os.listdir(directory)
        dx_values = sorted(set([float(re.search(f'{conc_value.replace(".", "_")}_dx_(\d+.\d+).p', file).group(1)) 
                                for file in filenames if re.search(f'{conc_value.replace(".", "_")}_dx_(\d+.\d+).p', file)]))

        m = m_values[conc_value]
        ka = ka_values[conc_value]
        h = h_values[conc_value]
        scaling_factor = scaling_factors[conc_value]

        K = np.power(time_to_steady_state, -1.0) * np.power(ka, -m) * np.power(1 - h * m, -1) * (
                np.power(L, 1 - h * m) - np.power(l, 1 - h * m))  # Calculate K
        K = K * scaling_factor

        U = Rf * K * np.power(ka, m) * (1 - h * m) * np.power((np.power(L, 1 - h * m) - np.power(l, 1 - h * m)), -1)
        ks = (U / K)

        all_std_values = []  # All data points
        lower_envelope_values = []  # Lower envelope points

        for dx in dx_values:
            file_path = os.path.join(directory, conc_value.replace('.', '_') + '_dx_' + str(dx) + '.p')
            with open(file_path, 'rb') as file:
                elev = p.load(file)
            
            grid = elev._griddata
            grid = np.pad(grid, ((1, 1), (0, 0)), 'constant', constant_values=0)
            m_size = elev._georef_info.ny + 2
            n_size = elev._georef_info.nx
            
            a, s = area_calc(grid, dx=dx)  
            ks_obs = np.reshape(np.power(a, m) * s, (m_size * n_size,))
            resid = ks_obs - ks
            std_ks_obs = np.std(resid) / ks
            all_std_values.append((dx, std_ks_obs))

        # Find the lower envelope
        lower_envelope_values = lower_envelope(all_std_values)

        # Store the results
        results[conc_value] = {dx: {'std': std} for dx, std in lower_envelope_values}

        # Plot all data (slightly transparent)
        if all_std_values:
            dx_values_all, std_values_all = zip(*all_std_values)
            plt.plot([1/dx for dx in dx_values_all], std_values_all, marker=markers[index % len(markers)],
                     linestyle='', color=colors[index], alpha=0.25, label=f'{conc_value} all data', markersize=3)

        # Plot lower envelope data (fully opaque)
        if lower_envelope_values:
            dx_values_env, std_values_env = zip(*lower_envelope_values)
            plt.plot([1/dx for dx in dx_values_env], std_values_env, marker=markers[index % len(markers)], 
                     linestyle='', color=colors[index], label=f'{conc_value} envelope data', markersize=3)

    # Plot formatting
    plt.xlabel('Inverse of dx')
    plt.ylabel('std(ks-obs - ks-mod)/ks-mod')
    plt.ylim(1E-4)
    plt.title('Inverse of dx vs Normalized Standard Deviation')
    plt.legend(loc='best')
    plt.tight_layout()
    plt.semilogy()
    plt.show()

    # Perform regression on lower envelope data and plot (use existing fit_and_evaluate_results function)
    ks_mod_dict = {
    'conc0.1': 0.42515966,  
    'conc0.3': 6.754379424601527,  
    'conc0.5': 88.97890120918466  
    }
    
    fit_parameters, r2_scores = fit_and_evaluate_results(results, ks_mod_dict)

    return results, fit_parameters, r2_scores

# Usage example:
conc_values = ['conc0.5'] # Use conc0.5, 0.3, or 0.1
results, fit_parameters, r2_scores = automated_process_and_plot(conc_values, directory)
```
